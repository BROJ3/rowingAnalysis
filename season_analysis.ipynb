```python
columns_of_interest = ['Log ID', 'Date', 'Description', 'Work Time (Formatted)',
       'Work Time (Seconds)', 'Rest Time (Formatted)', 'Rest Time (Seconds)',
       'Work Distance', 'Rest Distance', 'Stroke Rate/Cadence', 'Stroke Count',
       'Pace', 'Avg Watts', 'Cal/Hour', 'Total Cal', 'Avg Heart Rate',
       'Drag Factor']
```

- Get 4x20s and 7-5-2s
- plot split and HR over time
- store as database

```python
import sqlite3
```

```python
conn = sqlite3.connect('your_database.db')
```

```python
df.to_sql('ErgSeason', conn, index=False, if_exists='replace')
```

```python
# Execute an SQL query
query = 'SELECT * FROM ErgSeason WHERE Description = "4x20:00/5:00r row"'
result = pd.read_sql_query(query, conn)
```

```python
result
```

```python
# plot
px.line(x=result['Date'], y=result['Work Distance'], title='Distace covered Over Time')
```

```python
px.line(x=result['Date'], y=result['Pace'], title='Pace over time')
```

```python
fig = px.line(x=result['Date'], y=result['Avg Heart Rate'], title='Avg HR over time')

# Set the y-axis range manually
fig.update_layout(yaxis=dict(range=[0, 200]))

# Show the plot
fig.show()
```
